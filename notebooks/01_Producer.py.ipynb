### Step 2 — Simulate a Producer 
Re-run this to simulate a producer. NOTE: Should be run manually each time

```python
import json, random, uuid
from datetime import datetime, timezone
from pyspark.sql import Row
from builtins import round as python_round

categories = ["Electronics", "Clothing", "Groceries", "Sports", "Books"]

def generate_transactions(n=20):
    return [
        {
            "transaction_id": str(uuid.uuid4()),
            "event_time": datetime.now(timezone.utc).isoformat(),
            "category": random.choice(categories),
            "amount": python_round(random.uniform(10.0, 500.0), 2),
            "quantity": random.randint(1, 5)
        }
        for _ in range(n)
    ]

# Write a new batch file each time you run this
batch_id = str(uuid.uuid4())[:8]
file_path = f"/Volumes/streaming_demo/sales/raw_transactions/batch_{batch_id}.json"

transactions = generate_transactions(20)
json_lines = "\n".join(json.dumps(t) for t in transactions)

dbutils.fs.put(file_path, json_lines, overwrite=True)
print(f"✅ Written {len(transactions)} records to {file_path}")
```
